```python
#Dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# File Path
movies_file = "movies.csv"
# Read in the Data
df_movies = pd.read_csv(movies_file)
```

```python
df_movies.head()
```

```python
#Splitting the title and year 
movie_clean=pd.concat([df_movies,df_movies.title.str.split("(",expand=True)],1)
movie_clean.head()
```

```python
#Rename the new columns
movie_clean = movie_clean.rename(columns={0: "title_updated",
                                        1: "year_updated"})
movie_clean.head()
```

```python
#Splitting the title and year 
movie_clean=pd.concat([movie_clean,movie_clean.year_updated.str.split(")",expand=True)],1)
movie_clean.head()
```

```python
#Rename the new columns
movie_clean = movie_clean.rename(columns={0: "year"})
movie_clean.head()
```

```python
#Drop the columns I don't need
movie_updated_df = movie_clean[["movieId", "title", "title_updated", "year", "genres"]]
movie_updated_df.head()
```

```python
movie_updated_df.info()
#58,098 movies in the large database
```

```python
# Export file as a CSV
movie_updated_df.to_csv("movie_updated.csv", index=False, header=True)
```
